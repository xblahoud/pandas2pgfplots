```python
import pandas as pd
import numpy as np

from pandas2pgfplots import scatter, Plot
```

```python
df = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])
df.head()
```

```python
plot = scatter(df, diagonal=True)
plot
```
